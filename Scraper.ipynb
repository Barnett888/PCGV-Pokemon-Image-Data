Scraper For Pokemon images from official Pokemon website

```python
import cv2
import numpy as np
import urllib.request
import time
import os
```

```python
from os import getcwd
getcwd()
os.chdir('K:/Images/PKM_Mega') # set to the folder where you'd like to store mega pokemon images
```

```python
mega_dex_nums = ['003_f2','006_f2','006_f3','009_f2','015_f2','018_f2','065_f2','080_f2','094_f2','115_f2','127_f2','130_f2','142_f2',
'150_f2','150_f3','181_f2','208_f2','212_f2','214_f2','229_f2','248_f2','254_f2','257_f2','260_f2','282_f2','302_f2','303_f2', '306_f2','308_f2',
'310_f2','319_f2','323_f2','334_f2', '354_f2','359_f2','362_f2','373_f2','376_f2', '380_f2','381_f2','384_f2','428_f2','445_f2','448_f2',
'460_f2','475_f2','531_f2','719_f2',] 
```

Acquiring all Mega Evolutions

```python
start_time = time.time()
for i in range(0, len(mega_dex_nums)):
    try:
        req = urllib.request.Request(
            'https://assets.pokemon.com/assets/cms2/img/pokedex/full/' + mega_dex_nums[i] + '.png')
        response = urllib.request.urlopen(req)
        rr = response.read()
        ba = bytearray(rr)

        image = np.asarray(ba, dtype="uint8")
        image = cv2.imdecode(image, cv2.IMREAD_UNCHANGED)
        
        if os.path.isdir('Mega_' + mega_dex_nums[i]):
            os.chdir('Mega_' + mega_dex_nums[i])
            cv2.imwrite('Mega_' + mega_dex_nums[i] + ".png", image)
        else:
            os.mkdir('Mega_' + mega_dex_nums[i])
            os.chdir('Mega_' + mega_dex_nums[i])
            cv2.imwrite('Mega_' + mega_dex_nums[i] + ".png", image)
            
        os.chdir('K:/Images/PKM_Mega/')

        print("Saved " + mega_dex_nums[i] + ".png")

    except Exception as e:
        print("Error Occured for Pokemon " + '{:04d}'.format(i))
        print(str(e))
        
end_time = time.time()
print("Done")
print("Time Taken = ", end_time - start_time, "sec")
```
